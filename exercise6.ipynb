```python
import numpy as np
import matplotlib.pyplot as plt

x_coords = [0,1,2,3]
y_coords = [1.2,0.2,2.9,2.1]



t = np.linspace(0, 3)

y0 = 1.6+0*t
y1 = 0.79 + 0.54*t
y2 = 0.84 + 0.39*t + 0.05*t**2 
y3 = 1.2 - 5.25*t + 5.45*(t**2) - 1.2*(t**3)
y4 = 1.2 - 8.85*t + 10.25*(t**2) - 2.4*(t**3) - 1.63790491e-16*(t**4)

plt.plot(t,y0, label='y0')
plt.plot(t,y1, label='y1')
plt.plot(t, y2, label='y2')
plt.plot(t, y3, label='y3')
plt.plot(t, y4, label='y4')
plt.scatter(x_coords, y_coords)
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Values')
plt.grid(True)
plt.show()


```
